```python
# Imports
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Loading the results files

file_path_0 = 'DQN_continous_seed0.npy'  # modify here
file_path_1 = 'DQN_continous_seed1.npy'  # modify here
file_path_2 = 'DQN_continous_seed4long.npy'  # modify here
file_path_3 = 'DQN_continous_seed3.npy'  # modify here


# ---------------- SEED 0 --------------------

# Check if the file exists
if os.path.exists(file_path_0):
    # Load the array from a .npy file
    data0 = np.load(file_path_0)
    data0 = data0[:5920]
    #data0 = np.append(data0, np.load('DQN_continous_seed5.npy'))    # print(data0)
else:
    print(f"File not found: {file_path_0}. Check the file name and path.")


# ---------------- SEED 1 --------------------

# Check if the file exists
if os.path.exists(file_path_1):
    # Load the array from a .npy file
    data1 = np.load(file_path_1)
else:
    print(f"File not found: {file_path_1}. Check the file name and path.")


# # ---------------- SEED 2 --------------------

# Check if the file exists
if os.path.exists(file_path_2):
    # Load the array from a .npy file
    data2 = np.load(file_path_2)
    data2 = data2[:1000]

else:
    print(f"File not found: {file_path_2}. Check the file name and path.")

# # ---------------- SEED 3 --------------------
if os.path.exists(file_path_3):
    # Load the array from a .npy file
    data3 = np.load(file_path_3)
else:
    print(f"File not found: {file_path_3}. Check the file name and path.")
```

```python
# Plotting

# Generating x-axis values (timesteps)
timesteps = np.arange(0, len(data1))


# Plotting
plt.figure(figsize=(20,4))

#plt.plot(timesteps, data0, label='cold start (one seed because it took 16h)')
plt.plot(timesteps, data1, label='trained network')
plt.plot(timesteps, data2, label='first convergence')
plt.plot(timesteps, data3, label='first convergence')

plt.xlabel('Episode')
plt.ylabel('Episode Reward')
plt.title('DQN Mountain Car Continuous v0')
plt.legend()
plt.grid(True)
# Save the plot
#plt.savefig('CHANGETHENAMEHERE.png') #modify file name !!!!!!!!!!!!!!!!!!!!!

# Show the plot
plt.show()
```
