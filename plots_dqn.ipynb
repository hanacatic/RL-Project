```python
# Plotting

# Generating x-axis values (timesteps)
timesteps = np.arange(0, len(data1))


# Plotting
#plt.figure(figsize=(20,4))


# plt.plot(timesteps, data0, label='cold start (one seed because it took 16h)')
# plt.plot(timesteps, data0, label='seed 0 first convergence')
plt.plot(timesteps, data1, label='seed 1 first convergence')
plt.plot(timesteps, data2, label='seed 2 full convergence')
plt.plot(timesteps, data3, label='seed 3 first convergence')
plt.plot(timesteps, data4, label='seed 4 no convergence')
# plt.plot(timesteps, data5, label='seed 5 first convergence')
plt.plot(timesteps, data6, label='seed 6 no convergence')
plt.plot(timesteps, data7, label='seed 7 no convergence')
plt.plot(timesteps, data8, label='seed 8 first convergence')
plt.plot(timesteps, data9, label='seed 9 full convergence')
plt.plot(timesteps, data10, label='seed 10 first convergence')
plt.plot(timesteps, data11, label='seed 11 full convergence')



plt.xlabel('Episode')
plt.ylabel('Episode Reward')
plt.title('DQN Mountain Car Continuous v0')
plt.legend()
plt.grid(True)
# Save the plot
#plt.savefig('CHANGETHENAMEHERE.png') #modify file name !!!!!!!!!!!!!!!!!!!!!

# Show the plot
plt.show()
```

```python
# Calculating the average and standard deviation
average_data_no_conv = (data4 + data6 + data7) / 3
std_data_no_conv = np.std([data4, data6, data7], axis=0)

# Calculating the average and standard deviation
average_data_conv = (data1 + data3 + data8) / 3
std_data_conv = np.std([data1, data3, data8], axis=0) 

# Calculating the average and standard deviation
average_data_full = (data2 + data9 + data11) / 3
std_data_full = np.std([data2, data9, data11], axis=0)

# calculating the average and standard deviation TD3
average_data_td3 = (data31 + data32 + data33) / 3
std_data_td3 = np.std([data31, data32, data33], axis=0)
 
# cut the data to the same length
average_data_no_conv = average_data_no_conv[:len(average_data_td3)]
std_data_no_conv = std_data_no_conv[:len(average_data_td3)]

# Generating x-axis values (timesteps)  
timesteps = np.arange(0, len(average_data_no_conv))

# Plotting
plt.figure(figsize=(7, 5))
plt.plot(timesteps, average_data_no_conv, label='average_DQN', color='blue')
plt.fill_between(timesteps, average_data_no_conv - std_data_no_conv, average_data_no_conv + std_data_no_conv, color='cyan', alpha=0.3, label='standard deviation')
#plt.plot(timesteps, average_data_conv, label='average_first convergence', color='purple')
#plt.fill_between(timesteps, average_data_conv - std_data_conv, average_data_conv + std_data_conv, color='pink', alpha=0.3, label='standard deviation')
#plt.plot(timesteps, average_data_full, label='average_full convergence', color='green')
#plt.fill_between(timesteps, average_data_full - std_data_full, average_data_full + std_data_full, color='lime', alpha=0.3, label='standard deviation')
plt.plot(timesteps, average_data_td3, label='average_TD3', color='green')
plt.fill_between(timesteps, average_data_td3 - std_data_td3, average_data_td3 + std_data_td3, color='lime', alpha=0.3, label='standard deviation')
plt.xlabel('episode')
plt.ylabel('Episode Reward')
plt.title('Comparison of DQN and TD3 in Mountain Car Continuous v0')
plt.legend()
plt.grid(True)
# Save the plot
plt.savefig('plots/compare.png')
# Show the plot
plt.show()

```
